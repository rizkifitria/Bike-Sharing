## Exploratory Data Analysis (EDA)

Business Question :
- Bagaimana pengaruh musim terhadap frekuensi penyewaan sepeda?
- Pada jam berapa jumlah penyewaan sepeda paling tinggi dan paling sedikit?

### 1. Data day_df

```python
day_df.describe(include="all")
```

Berdasarkan rangkuman parameter di atas, kita akan memperoleh informasi jumlah penyewaan sepeda sebanyak 731 dalam 2 tahun dan menurut jenis penyewa per harinya.

```python
#frekuensi penyewaan perharinya menurut musim
day_df.season.value_counts()
```

```python
# Menggantikan angka dengan deskripsi
season_dict =  {
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
}
day_df['season'] = day_df['season'].map(season_dict)
```

```python
day_df.season.value_counts()
```

```python
day_df.groupby(by="season").agg({
    "casual" : ["max", "min", "mean"],
    "registered" : ["max", "min", "mean"],
    "cnt" : ["mean"]
})
```

Berdasarkan hasil tersebut, diketahui bahwa rata-rata penyewaan terbanyak terjadi pada musim gugur dan merupakan penyewa yang terdaftar.

```python
day_df.yr.value_counts()
```

```python
# Menggantikan angka dengan deskripsi
year_dict =  {
    0: '2011', 1: '2012'
}
day_df['yr'] = day_df['yr'].map(year_dict)
```

```python
day_df.groupby(by="yr").agg({
    "casual" : ["sum"],
    "registered" : ["sum"],
    "cnt" : ["sum"]
})
```

Berdasarkan hasil tersebut diketahui terjadi kenaikan jumlah penyewaan dari tahun sebelumnya.

```python
day_df.mnth.value_counts()
```

```python
# Menggantikan angka dengan deskripsi
month_dict =  {
    1: 'January', 2: 'February', 3: 'March',  4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
day_df['mnth'] = day_df['mnth'].map(month_dict)
```

```python
day_df.groupby(by="mnth").agg({
    "dteday" : "nunique",
    "casual" : "mean",
    "registered" : "mean",
    "cnt" : "mean"
}).sort_values(by="cnt", ascending=False)
```

Berdasarkan hasil di atas, jumlah penyewaan sepeda terbanyak terjadi pada bulan Juni dan paling rendah terjadi pada bulan Januari.

```python
day_df.workingday.value_counts()
```

```python
# Menggantikan angka dengan deskripsi
workingday_dict =  {
    0 : 'holiday', 1 : 'workingday'
}
day_df['workingday'] = day_df['workingday'].map(workingday_dict)
```

```python
day_df.groupby(by="workingday").agg({
    "dteday": "nunique",
    "casual" : "mean",
    "registered" : "mean",
    "cnt" : "mean"
})
```

```python
day_df.weekday.value_counts()
```

```python
# Menggantikan angka dengan deskripsi
weekday_dict =  {
    0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'
}
day_df['weekday'] = day_df['weekday'].map(weekday_dict)
```

```python
day_df.groupby(by="weekday").agg({
    "dteday": "nunique",
    "casual" : "mean",
    "registered" : "mean",
    "cnt" : "mean"
}).sort_values(by="cnt", ascending=False)
```

Berdasarkan hasil di atas, penyawaan paling tinggi rata-rata terjadi pada hari Sabtu dan paling rendah pada hari Senin. 

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh musim terhadap frekuensi penyewaan sepeda?

```python
day_df.rename(columns={
    "cnt" : "customer_count"
}, inplace=True)
byseason = day_df.groupby(by="season").customer_count.mean().reset_index().sort_values(by="customer_count", ascending=False)
byseason.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.figure(figsize=(10, 5))
colors1 = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(
    x="season",
    y="customer_count", 
    data=byseason.sort_values(by="customer_count", ascending=False),
    palette=colors1
)
plt.title("Jumlah Penyewaan Sepeda Berdasarkan Musim", loc="center", fontsize=15)
plt.xlabel('Musim')
plt.ylabel('Frek. Penyewaan Sepeda')
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Pertanyaan 2: Kapan terjadi penyewaan sepeda paling tinggi dan paling rendah?

```python
sum_hour = hour_df.groupby("hr").cnt.sum().sort_values(ascending=False).reset_index()
sum_hour.head(24)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 15))

colors=["#D3D3D3", "#D3D3D3", "#90CAF9", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="hr", y="cnt", data=sum_hour.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Waktu tingginya penyewaan ", loc="center", fontsize=20)
ax[0].tick_params(axis='y', labelsize=25)
ax[0].tick_params(axis='x', labelsize=20)

sns.barplot(x="hr", y="cnt", data=sum_hour.sort_values(by="hr", ascending=True).head(5), palette=["#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3","#90CAF9"], ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("Waktu terendah penyewaan", loc="center", fontsize=20)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].tick_params(axis='y', labelsize=25)
ax[1].tick_params(axis='x', labelsize=20)

plt.suptitle("Penyewaan Sepeda Tertinggi dan Terendah", fontsize=40)
plt.show()
```
